**EFECTO DE LA TEMPERATURA DE UNO DE LOS RESERVORIOS EN EL CAMBIO TOTAL DE ENTROPIA**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression 
#model 
```

```python
x = np.array([2,4,6,8])
y = np.array([5,15,37,63])
```

```python
linreg = LinearRegression()
x = x.reshape(-1, 1)
plt.figure(figsize=[10,6])
linreg.fit(x, y)
y_pred = linreg.predict(x)
plt.scatter(x, y, label = 'Datos experimentales')
plt.plot(x, y_pred,'--',  label = '$y = 200.24x$')
plt.xlabel('$T^{-1}$ $[K^-1]$', fontdict={'fontsize': 17})
plt.ylabel('$Entropia$ $[J/K]$', fontdict={'fontsize': 17})
axes = plt.gca()
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
#axes.set_xlim([0.005, 0])

first_legend = plt.legend(loc = 'upper right', shadow = True, fontsize=15)
ax = plt.gca().add_artist(first_legend)
plt.legend(['Regresion Lineal'], loc = 'lower right', shadow = True, fontsize=15)
#plt.savefig('AluminioLinealVSTemp.png', dpi = 300)
plt.show()
```

**EFECTO DEL INCREMENTO DEL CALOR TRANSFERIDO EN EL CAMBIO TOTAL DE LA ENTROPIA**

```python
x = np.array([200, 250, 300, 350, 400, 450, 500, 550, 600, 650, 700, 750, 800])
y = np.array([0.601763, 0.769077, 0.911877, 1.06168, 1.21744, 1.40636, 1.51451, 1.67751, 1.83944, 1.96386, 2.10546, 2.25518, 2.47366])

linreg = LinearRegression()
x = x.reshape(-1, 1)
plt.figure(figsize=[10,6])
linreg.fit(x, y)
y_pred = linreg.predict(x)
plt.scatter(x, y, label = 'Datos experimentales')
plt.plot(x, y_pred,'--',  label = '$y = 0.003x$')
plt.xlabel('$\Delta Q$ $[J]$', fontdict={'fontsize': 17})
plt.ylabel('$\Delta S$ $[J/K]$', fontdict={'fontsize': 17})
axes = plt.gca()
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)

first_legend = plt.legend(loc = 'upper left', shadow = True, fontsize=15)
ax = plt.gca().add_artist(first_legend)
plt.legend(['Regresion Lineal'], loc = 'lower right', shadow = True, fontsize=15)
plt.savefig('SvsQ', dpi = 300)
plt.show()
```
