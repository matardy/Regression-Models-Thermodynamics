```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#tiempo [s]     t_o = 27 [C]
x = np.array([30, 60, 90, 120, 150, 180 , 210, 240, 270, 300, 330, 360, 390, 420, 450, 480])
#variacion temperatura (V = 200 cm^3) [C]
y = np.array([2.56, 5.16, 7.56, 10.09, 12.61, 15.31, 17.58, 20.64, 22.88, 25.63, 28.12, 30.08, 32.81, 35.07, 38.6, 40.71])
#variacion temperatura (V = 300 cm^3) [C]
w = np.array([1.69, 3.36, 5.05, 6.8, 8.38, 10.24, 11.39, 13.61, 15.39, 17.11, 18.74, 20.21, 22.27, 23.73, 25.8, 27.41])
#masa x V = 200
w1 = np.array([0.512, 1.032, 1.512, 2.018, 2.522, 3.062, 3.516, 4.128, 4.576, 5.126, 5.624, 6.016, 6.562, 7.014, 7.72, 8.142])
#masa x V = 300
w2 = np.array([0.507, 1.008, 1.515, 2.04, 2.514, 3.072, 3.417, 4.083, 4.617, 5.133, 5.622, 6.063, 6.681, 7.128, 7.74, 8.223])
#Energia electrica de calentamiento
e = np.array([2.160, 4.320, 6.480, 8.640, 10.800, 12.960, 15.120, 17.280, 19.440, 21.600, 23.760, 25.920, 28.080, 30.240, 32.400, 34.560])
```

```python
#Energia electrica de calentamiento en funcion de la temperatura x masa
#V = 200 cm^3
linreg = LinearRegression()
w1 = w1.reshape(-1, 1)
plt.figure(figsize=[6,4.5])
linreg.fit(w1, e)
e_pred = linreg.predict(w1)
plt.scatter(w1, e, label = 'Datos experimentales')
plt.plot(w1, e_pred,'--', label = '$y = 4.2511$ $x$')
plt.xlabel('$ m \cdot \Delta T  $ $[kg \cdot ^{ยบ}C]$', fontdict={'fontsize': 14})
plt.ylabel('$ U $ $[mJ]$', fontdict={'fontsize': 20})
axes = plt.gca()
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
#axes.set_ylim([0,1099600])
first_legend = plt.legend(loc = 'upper left', shadow = True, fontsize='large')
ax = plt.gca().add_artist(first_legend)
plt.legend(['Regresion Lineal'], loc = 'lower right', shadow = True, fontsize=15)
plt.savefig('EnergiaVSTemp200.png', dpi = 200)
plt.show()
```

```python
print(linreg.coef_)
```

```python
print(linreg.intercept_)
```

```python
#Energia electrica de calentamiento en funcion de la temperatura x masa
# V = 300 [cm^3]
linreg = LinearRegression()
w2 = w2.reshape(-1, 1)
plt.figure(figsize=[6,4.5])
linreg.fit(w2, e)
e_pred = linreg.predict(w2)
plt.scatter(w2, e, label = 'Datos experimentales')
plt.plot(w2, e_pred,'--', label = '$y = 4.1943$ $x$')
plt.xlabel('$ m \cdot \Delta T  $ $[kg \cdot ^{ยบ}C]$', fontdict={'fontsize': 14})
plt.ylabel('$ U $ $[mJ]$', fontdict={'fontsize': 20})
axes = plt.gca()
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
#axes.set_ylim([0,1099600])
first_legend = plt.legend(loc = 'upper left', shadow = True, fontsize='large')
ax = plt.gca().add_artist(first_legend)
plt.legend(['Regresion Lineal'], loc = 'lower right', shadow = True, fontsize=15)
plt.savefig('EnergiaVsTemp300', dpi = 200)
plt.show()
```
