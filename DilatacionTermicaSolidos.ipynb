```python
import numpy as np # mumpy sirve para arrays
import matplotlib.pyplot as plt #Graficas
from sklearn.linear_model import LinearRegression #para hacer regresion lineal
```

```python
x = np.array([225, 250, 275, 300, 325, 350, 375, 400, 425, 450, 475])
y = np.array([1.00055, 1.00068, 1.00167, 1.00223, 1.00282, 1.00331, 1.00395, 1.00435, 1.00487, 1.0056, 1.00614])
z = np.array([1.00112, 1.0024, 1.00327, 1.00433, 1.00568, 1.0066, 1.00767, 1.00885, 1.01016, 1.01104, 1.01245])
w = np.array([1.00167, 1.00338, 1.00489, 1.00665, 1.00816, 1.01024, 1.01153, 1.01354, 1.01504, 1.01679, 1.01825])
```

```python
#Aluminio Lineal

linreg = LinearRegression() # Declaro un objeto de la libreria LinearRegression.  

x = x.reshape(-1, 1) #Ajusta el eje x para objetivos visuales

plt.figure(figsize=[10,6]) # Existe el cuadro del grafico

linreg.fit(x, y) #Aqui el modelo de regresion sabe que trabaj con x y y

y_pred = linreg.predict(x) # y = f(x)

#Plotting 

plt.scatter(x, y, label = 'Datos experimentales', color = 'pink')
plt.plot(x, y_pred,'--', label = 'y = 2.257x10^(-5) X + 09954 ',  color = 'red')
plt.xlabel('Tf [K]', fontdict={'fontsize': 15})
plt.ylabel('Lf [m]', fontdict={'fontsize': 15})
axes = plt.gca()
axes.set_ylim([1.0000,1.0075])
first_legend = plt.legend(loc = 'upper left', shadow = True, fontsize='large')
ax = plt.gca().add_artist(first_legend)
plt.legend(['Regresion Lineal'], loc = 'lower right', shadow = True, fontsize='large')
plt.savefig('AluminioLinealVSTemp.png', dpi = 300)
plt.show()
```

```python
print(linreg.coef_)
```
